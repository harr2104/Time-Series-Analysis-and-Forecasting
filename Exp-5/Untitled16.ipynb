```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
df = pd.read_csv("D:\\TSA\\Exp-1\\Dataset.csv")
df['month'] = pd.to_datetime(df['month'] + '-2020', format='%d-%b-%Y')
df.set_index('month', inplace=True)
ts = df['price'].dropna()
X = np.arange(len(ts)).reshape(-1, 1)
y = ts.values
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
plt.figure(figsize=(10, 5))
plt.plot(ts.index, y, label='Actual Data')
plt.plot(ts.index, y_pred, label='Linear Regression Forecast', color='red', linestyle='--')
plt.title('Linear Regression Forecasting')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
